# Анализ ассортимента обуви (Спорт-Марафон)

## Цель исследования

Провести комплексный анализ ассортимента обуви интернет-магазина «Спорт-Марафон»:

- изучить структуру ассортимента по брендам, типам обуви и категориям пола;
- проанализировать распределение цен и скидок;
- выявить особенности размерных рядов;
- проверить гипотезы о влиянии полноты размерного ряда и популярности брендов на глубину скидки.

## Данные

- Источник: сайт «Спорт-Марафон» (парсинг)
- Объем выборки: 1 395 моделей обуви (после очистки)
- Поля: бренд, название, цена, старая цена (для расчета скидки), доступные размеры

**Период сбора:** май 2026 г.

## Структура ассортимента

- **Всего брендов:** 48
- **Типы обуви:** кроссовки (53,5%), ботинки (35,7%), сандалии, сапоги, пуховые тапки
- **Распределение по полу:** мужские (45%), женские (39%), детские (8%), унисекс (8%)

## Цены

- **Средняя цена:** 19 978 руб.
- **Медианная цена:** 19 290 руб.
- **90% товаров:** от 5 248 до 35 990 руб.
- **Выбросы (более 39 278 руб):** 45 моделей (3,2% выборки)
